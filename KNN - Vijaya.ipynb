# Separate Dataset into Input & Output NumPy Arrays

The code separates the Pima diabetes dataset into input (X) and output (Y) arrays. X contains predictor variables (‘preg’, ‘plas’, ‘pres’, ‘skin’, ‘test’, ‘mass’, ‘pedi’, ‘age’), while Y contains the objective variable ‘class’. This data preparation step is important for training and testing machine learning models, as it separates features and target variables for making predictions

```python
# Step 3: Separate the dataset into input (X) and output (Y) arrays
X = df[['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']]
Y = df['class']  #target column is 'class'
```

# Split Input/Output Arrays into Training/Testing Datasets

The code divides the input (X) and output (Y) arrays into training and testing subsets. Of this, 67% of the data is used for training and 33% for testing. The random_state parameter with a value of 7 ensures that the same random split is repeatable in future runs, and provides consistency in model evaluation.

```python
# Split the dataset into training (67%) and testing (33%) subsets
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

# Build and Train the Model

The code below builds a K-Nearest Neighbors (KNN) classification model with variable ‘K’ neighbours. We are setting n_neighbors k_neighbors where we are choosing the value as 7, which means the KNN algorithm will consider all data points in the data set for classification. This means that the model will have a fixed number of neighbors, and any predictions will be based on all the data points in the training set. This rule is a K-Nearest Neighbors (KNN) classification model with a specified number of neighbors (7 in this case). The model will use all 7 nearest neighbors to predict additional data points based on their proximity in the feature space.

```python
# Train the KNN model with all neighbors
k_neighbors = 7  #choose the number of neighbors based on your preference
model = KNeighborsClassifier(n_neighbors=k_neighbors)
```

The code is training a K-Nearest Neighbors (KNN) classification model using the training data (X_train) and their corresponding traces (Y_train). For prediction accuracy, the model will use the given data to identify patterns and relationships between the factors and the target variable.

```python
k_neighbors = 7  # You can choose the number of neighbors based on your preference
model = KNeighborsClassifier(n_neighbors=k_neighbors)
# Train the model using the training data
model.fit(X_train, Y_train)
```

The code is making predictions about the testing process using the trained K-Nearest Neighbors (KNN) classification model. It then prints a classification report providing analytical metrics such as precision, recall, F1-score, and support for each subject in the test set as shown below. This report helps predict the outcome of the test data and assess the performance of the model. The code is a classification report on the performance of the KNN model on the test set. The report includes metrics such as accuracy, recall, and F1-score for each class (0 and 1) along with overall accuracy, weighted average, and weighted scores to help measure how well the model can classify each class sample and all in all well effort.

```python
# Predict the outcomes for the test set
predicted = model.predict(X_test)
# Print the classification report
report = classification_report(Y_test, predicted)
print("Classification Report: \n\n", report)
```

#Accuracy
The code below is calculating and printing the accuracy score of the KNN model on the test set. The precision score represents the proportion of samples that are best classified in the test set compared to the total number of samples.
```python
#score the accuracy level

result = model.score(X_test, Y_test)

#print out the results

print(("Accuracy: %.3f%%") % (result*100.0))
```

# Evaluate the model using the 10-fold cross-validation.
This code below uses the K-fold cross-validation method to evaluate the KNN model. It divides the data set into 10 clusters, stores random results for further generation, then trains a KNN model on each cluster while using the remaining clusters for validation The rule calculates the accuracy of the model per fold and then reports the mean and standard deviation as the result of a final analysis. The code outputs the evaluation results of the KNN model using 10-fold cross-validation. The average accuracy of the model is 0.721, and the standard deviation of the accuracy scores across the folds is 0.054. This indicates that the model has an average accuracy of approximately 72.1% and the accuracy scores have a small variation of 0.054, suggesting that the model's performance is relatively consistent across different folds.

```python
# evaluate the algorythm
# specify the number of time of repeated splitting, in this case 10 folds
n_splits = 10

# fix the random seed
# must use the same seed value so that the same subsets can be obtained
# for each time the process is repeated
seed = 7

# split the whole dataset into folds
# In k-fold cross-validation, the original sample is randomly partitioned into k equal sized
#subsamples. Of the k subsamples, a single subsample is retained as the validation data for
#testing the model, and the remaining k − 1 subsamples are used as training data. The crossvalidation process is the#n repeated k times, with each of the k subsamples used exactly once as
#the validation data. The k results can then be averaged to produce a single estimation. The
#advantage of this method over repeated random sub-sampling is that all observations are used for
#both training and validation, and each observation is used for validation exactly once
kfold = KFold(n_splits, random_state=seed, shuffle=True)

# for logistic regression, we can use the accuracy level to evaluate the model / algorithm
scoring = 'accuracy'

# train the model and run K-fold cross validation to validate / evaluate the model
results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)

# print the evaluationm results
# result: the average of all the results obtained from the K-fold cross validation
print("Accuracy: %.3f (%.3f)" % (results.mean(), results.std()))
```
